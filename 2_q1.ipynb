## Ques 1. Simulate NAND gate and three input AND gate using single/multi-layer perceptron.

```python
import tensorflow as tf
import numpy as np
```

####  Initializing variable for 2 Inputs

```python
X1=tf.Variable(initial_value=[0.,1.,0.,1.])
X2=tf.Variable(initial_value=[0.,0.,1.,1.])
Y_NAND=tf.Variable(initial_value=[1.,1.,1.,0.])
```

```python
class Model(object):

   def __init__(self):
        self.W1 = tf.Variable(2.)
        self.W2= tf.Variable(2.)
        self.b=tf.Variable(2.)

   def __call__(self, x1,x2):
        self.K1=self.W1 * x1 + self.W2 * x2 + self.b
        self.Output = tf.keras.activations.sigmoid(self.K1)
        return(self.Output)
model=Model()
```

```python
def compute_cost(target_y, predicted_y):
    c=tf.keras.losses.BinaryCrossentropy()
    return(c(target_y,predicted_y))
```

```python
def None_to_Zero(v):
    if v==None:
        v=0
    return v
```

```python
def train(model, X1, X2, Y2, learning_rate):
    with tf.GradientTape() as t:
            current_loss = compute_cost(Y2,model(X1,X2))
    dW1,dW2,db=t.gradient(current_loss,[model.W1,model.W2,model.b])
    dW1=None_to_Zero(dW1)
    db=None_to_Zero(db)
    dW2=None_to_Zero(dW2)
    model.W1.assign_sub(learning_rate * dW1)
    model.b.assign_sub(learning_rate * db)
    model.W2.assign_sub(learning_rate * dW2)
```

#### Model for 2 Input NAND

```python
# As we have very less Input data we need to increase the number of # epochs or the learning rate.
epochs=range(1000)
learning_rate=0.1
for epoch in epochs:
    train(model, X1, X2,Y_NAND, learning_rate)
#Once the training is done we can find results by calling the Model  #Class object as
print(np.round(model(X1, X2).numpy()))
```

#### Initializing variable for 3 Input

```python
X1=tf.Variable(initial_value=[0.,1.,0.,1.,0.,1.,0.,1.])
X2=tf.Variable(initial_value=[0.,0.,1.,1.,0.,0.,1.,1.])
X3=tf.Variable(initial_value=[0.,0.,0.,0.,1.,1.,1.,1.])

Y=tf.Variable(initial_value=[0.,0.,0.,0.,0.,0.,0.,1.])
```

```python
class Model1(object):

   def __init__(self):

         # Initialize the weights to `2.0` and the bias to `2.0`
         # In practice, these should be initialized to random values
         # self.W1=tf.random.normal([1])
        self.W1 = tf.Variable(2.)
        self.W2= tf.Variable(2.)
        self.W3= tf.Variable(2.)
        self.b=tf.Variable(2.)

   def __call__(self, x1,x2,x3):
        self.K1=self.W1 * x1 + self.W2 * x2 +self.W3 * x3+ self.b
        self.Output = tf.keras.activations.sigmoid(self.K1)
        return(self.Output)
model=Model1()
```

```python
def train(model, X1, X2, X3, Y2, learning_rate):
    with tf.GradientTape() as t:
            current_loss = compute_cost(Y2,model(X1,X2,X3))
    dW1,dW2,dW3,db=t.gradient(current_loss,[model.W1,model.W2,model.W3,model.b])
    dW1=None_to_Zero(dW1)
    db=None_to_Zero(db)
    dW2=None_to_Zero(dW2)
    dW3=None_to_Zero(dW3)
    model.W1.assign_sub(learning_rate * dW1)
    model.b.assign_sub(learning_rate * db)
    model.W2.assign_sub(learning_rate * dW2)
    model.W3.assign_sub(learning_rate * dW3)
```

#### Model for AND Gate

```python
# As we have very less Input data we need to increase the number of # epochs or the learning rate.
epochs=range(1000)
learning_rate=0.1
for epoch in epochs:
    train(model, X1, X2,X3,Y, learning_rate)
#Once the training is done we can find results by calling the Model  #Class object as
print(np.round(model(X1, X2,X3).numpy()))
```
